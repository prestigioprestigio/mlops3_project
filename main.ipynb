```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("./data/census.csv")
```

```python
print(df.shape)
display(df.head())
```

```python
for col in df.columns:
    if col[0] == ' ':
        df.rename(columns={col: col[1:]}, inplace=True)
```

```python
df.age.hist(bins=99, density=True)
```

```python
def plot_cat_col(col_name):
    fig, ax = plt.subplots(nrows=1, ncols=1)
    _ = ax.bar(df[col_name].unique(), df[col_name].value_counts()[df[col_name].unique()])
    _ = ax.set_xticklabels(df[col_name].unique(), rotation=90)
```

```python
cat_cols = ['workclass',
 'education',
 'education-num',
 'marital-status',
 'occupation',
 'relationship',
 'race',
 'sex',
 'native-country',
 'salary']
num_cols = ['age',
 'fnlgt',
 'capital-gain',
 'capital-loss',
 'hours-per-week',]
```

```python
for col in cat_cols:
    plot_cat_col(col)
```
